```python
train_data_new = train_data_num_features.join(nominal_attr)
```

```python
test_data_new.shape
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer()
train_data_new = imp.fit_transform(train_data_new)
test_data_new = imp.fit_transform(test_data_new)
```

```python
x_train,x_test,y_train,y_test=train_test_split(train_data_new, Y, test_size=0.2, random_state = 42)
```

```python
from sklearn.preprocessing import MinMaxScaler
s=MinMaxScaler(feature_range=(0,1))
```

```python
s.fit(x_train)
scaled_train=s.transform(x_train)
scaled_train
```

```python
scaled_test=s.transform(x_test)
scaled_test
```

```python
scaled_test_data=s.transform(test_data_new)
scaled_test_data
```

```python
from sklearn.ensemble import GradientBoostingRegressor
#from sklearn.model_selection import GridSearchCV
#model=GradientBoostingRegressor()
#params={'n_estimators':range(1,200),'max_depth': range(1,5)  }
#grid=GridSearchCV(estimator=model,cv=2,param_grid=params,scoring='neg_mean_squared_error')
#grid.fit(scaled_train,y_train)
#grid.best_estimator_
reg = GradientBoostingRegressor(learning_rate = 0.01, n_estimators = 100, max_depth = 1)
reg.fit(scaled_train,y_train)
```

```python
import xgboost as xgb
from xgboost import XGBRegressor
xgbr = XGBRegressor(learning_rate = 0.008, n_estimators = 250, max_depth = 1)
xgbr.fit(scaled_train,y_train)
```

```python
preds_train = xgbr.predict(scaled_train)
preds_test = xgbr.predict(scaled_test)
preds = list(preds_train) + (list(preds_test))
train_data["Predicted_SalePrice"] = preds
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print(mean_squared_error(Y,train_data["Predicted_SalePrice"]))
```
